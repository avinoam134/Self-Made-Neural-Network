```python
import Apply_Tests_For_Reports as ts
import matplotlib.pyplot as plt
from Neural_Layers import activations, Layer, LossLayer, ResidualLayer
from Neural_Networks import NeuralNetwork
```

```python
Yt, Ct, Yv, Cv = ts.parse_data("PeaksData")
Yt[:,0]
```

```python
input_dim = 2
hidden_layer = 6
num_samples= 32
relu_activation = activations["relu"]
layer1 = ResidualLayer(input_dim, num_samples ,relu_activation)
layer1_1 = Layer(input_dim, hidden_layer, relu_activation)
layer2 = ResidualLayer(hidden_layer, num_samples ,relu_activation)
layer2_1 = Layer(hidden_layer, input_dim, relu_activation)
layer3 = ResidualLayer(input_dim, num_samples ,relu_activation)
#layer3 = ResidualLayer(input_dim, num_samples ,tanh_activation)
loss_layer = LossLayer(input_dim, input_dim)
network = NeuralNetwork([layer1, layer1_1, layer2, layer2_1, layer3, loss_layer])
train_losses, test_loss, test_accurcy = ts.test_network_on_data_set("SwissRollData", network)
```

1. RELU brings better and faster results than tanh
2. Resnet bring much slower results (since W's have to be of size (num_classes, num_samples))
3. regular (non-res) nets throw division by zero errors

```python
# plot the losses and accuracies:
plt.plot(train_losses, label="train_loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
print ("test_loss: ", test_loss, "test_accurcy: ", test_accurcy)
plt.legend()
plt.show()
```
